```python
# Dependencies and Setup
import pandas as pd 
from pathlib import Path
import warnings
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

# Pull CSV
revenues_df = pd.read_csv("Resources/master_monthly_revenue.csv")
revenues_df.head()
```

```python
reverse_rev = revenues_df.iloc[::-1]
reverse_rev
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(data = reverse_rev,
            x = 'Month',
            y = 'Cumulative Gross',
            hue = 'Year',
            ci = None,
            color = 'salmon'
            )
plt.rcParams["figure.figsize"] = (1500, 600)
plt.show()
```
