```python
import pandas as pd
import requests


# Set target URLS for cleaning
url_2019 = 'https://www.boxofficemojo.com/month/by-year/2019/?grossesOption=calendarGrosses'
url_2020 = 'https://www.boxofficemojo.com/month/by-year/2020/?grossesOption=calendarGrosses'
url_2021 = 'https://www.boxofficemojo.com/month/by-year/2021/?grossesOption=calendarGrosses'
url_2022 = 'https://www.boxofficemojo.com/month/by-year/2022/?grossesOption=calendarGrosses'
url_2023 = 'https://www.boxofficemojo.com/month/by-year/2023/?grossesOption=calendarGrosses'
```

```python
# Grab 2019 table and put into Pandas
response = requests.get(url_2019)
table_19 = pd.read_html(response.text)
table_2019 = table_19[0]
table_2019
```

```python
# Clean table for merging

# Drop columns we won't be looking at
cleaned_2019 = table_2019[["Month", "Cumulative Gross", "% of Year", 'Releases','Average']]
cleaned_2019
```

```python
cleaned_2019.columns = ['Month','Cumulative Gross (2019)','% of Year (2019)','Releases (2019)','Average (2019)']
cleaned_2019
```

```python
# Grab 2020 table and put into Pandas
response = requests.get(url_2020)
table_20 = pd.read_html(response.text)
table_2020 = table_20[0]
table_2020
```

```python
# Drop columns we won't be looking at
cleaned_2020 = table_2020[["Month", "Cumulative Gross", "% of Year", 'Releases','Average']]
cleaned_2020
```

```python
cleaned_2020.columns = ['Month','Cumulative Gross (2020)','% of Year (2020)','Releases (2020)','Average (2020)']
cleaned_2020
```

```python
# Grab 2021 table and put into Pandas
response = requests.get(url_2021)
table_21 = pd.read_html(response.text)
table_2021 = table_21[0]
table_2021
```

```python
# Drop columns we won't be looking at
cleaned_2021 = table_2021[["Month", "Cumulative Gross", "% of Year", 'Releases','Average']]
cleaned_2021
```

```python
cleaned_2021.columns = ['Month','Cumulative Gross (2021)','% of Year (2021)','Releases (2021)','Average (2021)']
cleaned_2021
```

```python
# Grab 2022 table and put into Pandas
response = requests.get(url_2022)
table_22 = pd.read_html(response.text)
table_2022 = table_22[0]
table_2022
```

```python
# Drop columns we won't be looking at
cleaned_2022 = table_2022[["Month", "Cumulative Gross", "% of Year", 'Releases','Average']]
cleaned_2022
```

```python
cleaned_2022.columns = ['Month','Cumulative Gross (2022)','% of Year (2022)','Releases (2022)','Average (2022)']
cleaned_2022
```

```python
# Grab 2023 table and put into Pandas
response = requests.get(url_2023)
table_23 = pd.read_html(response.text)
table_2023 = table_23[0]
table_2023
```

```python
# Drop columns we won't be looking at
cleaned_2023 = table_2023[["Month", "Cumulative Gross", "% of Year", 'Releases','Average']]
cleaned_2023
```

```python
cleaned_2023.columns = ['Month','Cumulative Gross (2023)','% of Year (2023)','Releases (2023)','Average (2023)']
cleaned_2023
```

```python
master_df = cleaned_2019.merge(cleaned_2020, left_on = 'Month', right_on = 'Month')
master_df
```

```python
master_df = master_df.merge(cleaned_2021, left_on = 'Month', right_on = 'Month')
master_df
```

```python
master_df = master_df.merge(cleaned_2022, left_on = 'Month', right_on = 'Month')
master_df
```

```python
master_df = master_df.merge(cleaned_2023, left_on = 'Month', right_on = 'Month')
master_df
```

```python
from pathlib import Path

filepath = Path('Resources/master_monthly_revenue.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
master_df.to_csv(filepath, index=False)
```
